## Analysis of VisionSpeech embeddings

```python
# keywords_filter = ["baby", "beach", "bike", "climbing", "running", "tennis"]
# keywords_filter = [
#     "baby", "boy", "children", "young",
#     "ball", "field", "football", "grass", "running", "soccer"
#     ]
# keywords_filter = [
#     "dirt", "bike", "riding", "skateboard", "jumps", "ocean", "beach", "road", "street"
#     ]
keywords_filter = [
    "soccer", "football", "ball", "rides", "riding", "bike", "jumps", "air", "road", "street"
    ]
n_max_keywords = 50

# random.seed(3)
# keywords_filter = random.sample(keywords, 10)
# print("Keywords:", keywords_filter)
```

```python
# Get embeddings
# pkl_fn = "models/train_visionspeech_cnn/332147c538/sigmoid_output_dict.dev_queries_all.pkl" # D=1000
# pkl_fn = "models/train_visionspeech_cnn/332147c538/sigmoid_final_feedforward_dict.dev_queries_all.pkl" # D=1000
# pkl_fn = "models/train_visionspeech_cnn/332147c538/sigmoid_final_feedforward_dict.dev_queries.pkl" # D=3000
# pkl_fn = "models/train_visionspeech_cnn/18ba6618ad/sigmoid_final_feedforward_dict.dev_queries.pkl" # D=256
# pkl_fn = "models/train_visionspeech_cnn/18ba6618ad/sigmoid_final_feedforward_dict.dev_queries_all.pkl" # D=256
pkl_fn = "models/train_bow_cnn/12597afba4/sigmoid_final_feedforward_dict.dev_queries_all.pkl" # D=256
with open(pkl_fn, "rb") as f:
    features_dict = pickle.load(f)
labels = []
embeddings = []
counter = Counter()
for utt_key in features_dict:
    label = utt_key.split("_")[0]
    if label not in keywords_filter:
        continue
    if counter[label] > n_max_keywords:
        continue
    labels.append(label)
    counter[label] += 1
    embeddings.append(features_dict[utt_key])
```

```python
# Embedding information
embeddings = np.array(embeddings)
print("Embeddings shape:", embeddings.shape)
print("Embeddings min:", np.min(embeddings))
print("Embeddings max:", np.max(embeddings))
print("Embeddings mean:", np.mean(embeddings))
```

```python
tsne_perplexity = 30
tsne = manifold.TSNE(n_components=2, perplexity=tsne_perplexity, init="random", random_state=0)
X_tsne = tsne.fit_transform(embeddings)
```

```python
def plot_labelled_2d_data(X, labels, ncol=None):
    
    # Plot the different classes
    classes = set(labels)
    for label in classes:
        indices = np.where(np.array(labels) == label)[0]
        plt.scatter(X[indices, 0], X[indices, 1], label=label) #, c=np.array([label_to_color[label]]))
    if ncol is not None:
        plt.legend(ncol=ncol)
    else:
        plt.legend()
    plt.axis("off")

plot_labelled_2d_data(X_tsne, labels)
```

```python
# Get embeddings
pkl_fn = "models/train_visionspeech_cnn/18ba6618ad/sigmoid_final_feedforward_dict.dev_queries_all.pkl" # D=256
with open(pkl_fn, "rb") as f:
    features_dict = pickle.load(f)
labels = []
embeddings = []
counter = Counter()
for utt_key in features_dict:
    label = utt_key.split("_")[0]
    if label not in keywords_filter:
        continue
    if counter[label] > n_max_keywords:
        continue
    labels.append(label)
    counter[label] += 1
    embeddings.append(features_dict[utt_key])

# Embedding information
embeddings = np.array(embeddings)
print("Embeddings shape:", embeddings.shape)
print("Embeddings min:", np.min(embeddings))
print("Embeddings max:", np.max(embeddings))
print("Embeddings mean:", np.mean(embeddings))
```

```python
# Plot t-SNE embeddings
tsne = manifold.TSNE(n_components=2, perplexity=30, init="random", random_state=0)
X_tsne = tsne.fit_transform(embeddings)
plot_labelled_2d_data(X_tsne, labels, ncol=2)
```
